```python
import matplotlib.pyplot as plt

def plot_burgers(np_data, t_idx, num_x=500):
    """Plot the burgers equation solution.
    
    Args:
        np_data (np.ndarray): Solution to burgers equation.
        t_idx (int): Index of time to plot.
        save (bool): Whether to save the figure.
        save_path (str): Path to save the figure.
    """
    plt.figure(figsize=(10, 6))
    plt.plot(np_data[t_idx*num_x:(t_idx+1)*num_x, 1], np_data[t_idx*num_x:(t_idx+1)*num_x, 2], label='Numerical')
    plt.xlabel('x')
    plt.ylabel('u')
    plt.title(f'Burgers Equation Solution at t={np_data[t_idx*num_x, 0]:.2f}')
    plt.legend()
    plt.show()
```

```python
DATA_PATH = "/workspaces/pytorch-PINN/data/center_riemann_burgers.npy"
np_data = np.load(DATA_PATH)
```

```python
plot_burgers(np_data, 0)
```
